# Exercise 3: Lists
Goal: Practice working with lists, including accessing elements, calculating sums, averages, min,
and max, and modifying the list.

```python
# Create a list named monthly_expenses with values [2500, 3000, 2800, 3200, 2900].
monthly_expenses = [2500, 3000, 2800, 3200, 2900]
```

```python
# Calculate the total_expenses.
total_expenses = sum(monthly_expenses)
```

```python
#Calculate the average_expense.
average_expense = total_expenses / len(monthly_expenses)
```

```python
# Find the minimum_expense.
# Find the maximum_expense.
minimum_expense = min(monthly_expenses)
maximum_expense = max(monthly_expenses)
```

```python
# Print the total_expenses, average_expense, minimum_expense, and maximum_expense
print(f"Total Expenses: {total_expenses}")
print(f"Average Expense: {average_expense}")
print(f"Minimum Expense: {minimum_expense}")
print(f"Maximum Expense: {maximum_expense}")
```

```python
# Add a new expense value, 3100, to the monthly_expenses list using the .append() method.
monthly_expenses.append(3100)
```

```python
# Recalculate and print the new total_expenses and average_expense.
total_expenses = sum(monthly_expenses)
average_expense = total_expenses / len(monthly_expenses)
print(f"New Total Expenses: {total_expenses}")
print(f"New Average Expense: {average_expense}")
```

# Exercise 5: Conditionals
Goal: Practice using if, elif, and else statements within a function.

```python
# Define a function called check_discount that takes one argument, purchase_amount.
# Inside the function, write the following conditional logic:
# if purchase_amount &gt; 500:
# print(&quot;Eligible for 20% discount&quot;)
# elif purchase_amount &gt; 100:
# print(&quot;Eligible for 10% discount&quot;)
# else:
# print(&quot;No discount&quot;)
# Call the check_discount function three times with different values for purchase_amount to test all three conditions.
def check_discount(purchase_amount):
    if purchase_amount > 500:
        print("Eligible for 20% discount")
    elif purchase_amount > 100:
        print("Eligible for 10% discount")
    else:
      print("No discount")

check_discount(600)
check_discount(90)
check_discount(300)
```

# Exercise 7: Pandas DataFrame Creation and Manipulation
Goal: Create a Pandas DataFrame and perform basic manipulations.


```python
# Import the pandas library as pd.
import pandas as pd
```

```python
# Create a Python dictionary named student_data with the following keys and lists as values:
# &#39;Name&#39;: [&#39;Alice&#39;, &#39;Bob&#39;, &#39;Charlie&#39;, &#39;David&#39;]
# &#39;Math_Score&#39;: [85, 92, 78, 88]
# &#39;Science_Score&#39;: [90, 88, 85, 95]
# &#39;English_Score&#39;: [75, 80, 82, 79]
student_data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Math_Score': [85, 92, 78, 88],
    'Science_Score': [90, 88, 85, 95],
    'English_Score': [75, 80, 82, 79]
}
```

```python
# Create a Pandas DataFrame named students_df from the student_data dictionary.
students_df = pd.DataFrame(student_data)
```

```python
# Print the entire DataFrame.
print(students_df)
```

```python
# Add a new column named &#39;Average_Score&#39; to students_df by calculating the average of
# &#39;Math_Score&#39;, &#39;Science_Score&#39;, and &#39;English_Score&#39;.
students_df['Average_Score'] = students_df[['Math_Score', 'Science_Score', 'English_Score']].mean(axis=1)
```

```python
print(students_df)
```

```python
# Filter the students_df to show only students whose &#39;Average_Score&#39; is greater than 85 and print the filtered DataFrame.
high_achievers = students_df[students_df['Average_Score'] > 85]
print(high_achievers)
```

# Exercise 8: NumPy Array Operations
Goal: Create NumPy arrays and perform basic mathematical and conditional operations.

```python
# import the numpy library as np.
import numpy as np
```

```python
# Create two NumPy arrays:
# product_a_sales = np.array([120, 150, 130, 180, 160])
# product_b_sales = np.array([100, 130, 140, 170, 150])
product_a_sales = np.array([120, 150, 130, 180, 160])
product_b_sales = np.array([100, 130, 140, 170, 150])
```

```python
# Calculate the total_sales_per_month by adding the two arrays element-wise. Print the result.
# Calculate the difference_in_sales (Product A sales - Product B sales) element-wise. Print the result.
total_sales_per_month = product_a_sales + product_b_sales
print(f"Total Sales Per Month: {total_sales_per_month}")
difference_in_sales = product_a_sales - product_b_sales
print(f"Difference in Sales: {difference_in_sales}")

```

```python
# Calculate the average_sales_a and average_sales_b using np.mean(). Print both averages.
average_sales_a = np.mean(product_a_sales)
average_sales_b = np.mean(product_b_sales)
print(f"Average Sales for Product A: {average_sales_a}")
print(f"Average Sales for Product B: {average_sales_b}")
```

```python
# Create a boolean array is_a_greater that is True where product_a_sales is greater than
# product_b_sales. Print this boolean array.
is_a_greater = product_a_sales > product_b_sales
print("Is Product A's sales greater than Product B's each month?")
print(is_a_greater)
```

```python
# Use np.sum() on the is_a_greater array to count how many months Product A sales were greater
# than Product B sales. Print the count.
count_a_greater = np.sum(is_a_greater)
print(f"Number of months where Product A's sales were greater than Product B's: {count_a_greater}")
```

# Exercise 10: Seaborn Basic Plotting

```python
# Import the seaborn library as sns and matplotlib.pyplot as plt.
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Create a Pandas DataFrame category_data from a dictionary:
# &#39;Category&#39;: [&#39;Type A&#39;, &#39;Type B&#39;, &#39;Type A&#39;, &#39;Type C&#39;, &#39;Type B&#39;, &#39;Type C&#39;]
# &#39;Value&#39;: [10, 15, 12, 8, 18, 10]
data = {
    'Category': ['Type A', 'Type B', 'Type A', 'Type C', 'Type B', 'Type C'],
    'Value': [10, 15, 12, 8, 18, 10]
}

category_data = pd.DataFrame(data)

print(category_data)
```

```python
# Use sns.countplot() to create a count plot of the &#39;Category&#39; column. Add a title and display the plot.
sns.countplot(data=category_data, x='Category')
plt.title('Count Plot of Categories')
plt.show();
```

```python
#Use sns.barplot() to show the average &#39;Value&#39; for each &#39;Category&#39;. Add a title and display the plot.
sns.barplot(data=category_data, x='Category', y='Value')
plt.title('Average Value by Category')
plt.show();
```
